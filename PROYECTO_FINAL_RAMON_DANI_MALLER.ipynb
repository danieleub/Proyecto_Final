#### Paso 5: Construye el modelo y optimízalo

```python
feature_columns=['año', 'mes', 'dia', 'departamento', 'edad', 'sexo_F', 'sexo_M', 'fuente_tipo_contagio']

X=data_c[feature_columns]
Y=data_c[['recuperado']]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20)
```

##### Decision Tree Classifier

```python
modelo_DT = DecisionTreeClassifier()
modelo_DT.fit(X_train, y_train)
```

```python
y_pred = modelo_DT.predict(X_test)

accuracy_DT=metrics.accuracy_score(y_test, y_pred)
precision_DT = precision_score(y_test, y_pred, average='weighted')
recall_DT = recall_score(y_test, y_pred, average='weighted')
f1_DT = f1_score(y_test, y_pred, average='weighted')

print(f"Accuracy: {accuracy_DT}")
print(f"Precision: {precision_DT}")
print(f"Recall: {recall_DT}")
print(f"F1 Score: {f1_DT}")
```

```python
hyperparams = {
    "criterion": ["gini", "entropy"],
    "max_depth": [None, 5, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

grid = GridSearchCV(modelo_DT, hyperparams, scoring = "accuracy", cv = 10)
grid
```

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

grid.fit(X_train, y_train)

print(f"Best hyperparameters: {grid.best_params_}")
```

```python
modelo_DT = DecisionTreeClassifier(criterion = "entropy", max_depth = 5, min_samples_leaf = 4, min_samples_split = 2, random_state = 42)
modelo_DT.fit(X_train, y_train)
```

```python
y_pred = modelo_DT.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
joblib.dump(modelo_DT, 'modelo_DT.pkl')
```

##### Random Forest Classifier

```python
modelo_RF = RandomForestClassifier(n_estimators = 30, random_state = 20)
modelo_RF.fit(X_train, y_train)
```

```python
y_pred = modelo_RF.predict(X_test)

accuracy_RF = accuracy_score(y_test, y_pred)
precision_RF = precision_score(y_test, y_pred, average='weighted')
recall_RF = recall_score(y_test, y_pred, average='weighted')
f1_RF = f1_score(y_test, y_pred, average='weighted')

print(f"Accuracy: {accuracy_RF}")
print(f"Precision: {precision_RF}")
print(f"Recall: {recall_RF}")
print(f"F1 Score: {f1_RF}")
```

```python
estimator = modelo_RF.estimators_[0]

plt.figure(figsize=(20,10))
plot_tree(estimator,
          feature_names = list(X_train.columns),
          class_names = ["0", "1", "2"],
          filled = True,
          fontsize=8,
          max_depth=3) # Limit depth for better visualization
plt.title("Visualizacion del primer arbol de Random Forest")
plt.show()
```
